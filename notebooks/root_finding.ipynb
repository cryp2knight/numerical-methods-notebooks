```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Graphical methods
### example 1

```python
#function for solving the drag coefficient of the parachutist problem
def drag_coef(c):
    mass = 68.1 #kg
    velocity = 40 #m/s
    t = 10 #s
    acceleration_gravity = 9.81 #m/s^2
    return (acceleration_gravity*mass)/c*(1 - math.pow(math.e, -c/mass*t))- velocity
```

```python
vals = np.array(list(map(drag_coef, range(1,20))))
plt.plot(vals)
plt.grid(True, which='both')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
vals
```

### example 2

```python
# f(x)= sin 10x + cos 3x
def f(x):
    return np.sin(10*x)+np.cos(3*x)
```

```python
fig = plt.figure()
g1 = fig.add_subplot(2,2,1)
x1 = np.arange(4.2,4.3,0.001)
plt.xlim(4.2,4.3)
g1.plot(x1, f(x1))
plt.grid(True, which='both')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

g2 = fig.add_subplot(2,2,2)
x2 = np.arange(3,5,0.001)
plt.xlim(3,5)
g2.plot(x2, f(x2))
plt.grid(True, which='both')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

g3 = fig.add_subplot(2,2,3)
x3 = np.arange(0,5,0.01)
plt.xlim(0,5)
g3.plot(x3, f(x3))
plt.grid(True, which='both')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show()
```

### Bisection Method

```python
def f(x): #input your function here in terms of x
    return drag_coef(x) #using the first function to compute the drag_coefficient

pp = []
def bisection_method(u, l):
    while True:
        xr = (u + l)/2 #estimated root
        checker = f(l)*f(xr)
        pp.append(xr)
#         print(checker)
        if checker < 0: u = xr
        elif checker > 0: l = xr
        else: return xr

bisection_method(14,16) #estimated range for the root
```
